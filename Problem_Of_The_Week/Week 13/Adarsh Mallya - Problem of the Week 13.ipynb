## Problem 13.2 - k-Nearest Neighbors vs k-Mean Clustering

Both of the algorithms k-Nearest Neighbors vs k-Mean Clustering use the Euclidean distance to compare similarities between data points. What is the key difference between these algorithms? Which one is unsupervised and which one is supervised? Explain in 2-4 sentences.

k-Nearest Neighbors generates classifications based on the k nearest points to the current point, determining the label based on the majority classification of those k points. k-Mean Clustering generates classifications based on distance away from k centroid points of the distribution, with each point being assigned to the nearest centroid. K-Nearest Neighbors is supervised, as the model learns from existing classifications to predict the next one, while K-Means clustering is unsupervised since the model does not know any classifications beforehand and must classify all of them on its own.
