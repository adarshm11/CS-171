A neural network is a complex machine learning model that is used for making predictions in a similar fashion that neurons in the human brain do (hence the name "neural" network). In a neural network, calculations are made in nodes of the model called "neurons", and these layers of neurons are passed to successive layers, with each layer contributing toward the final result.

## Problem 11.2 - Model Purposes

For each of the NNs listed above, explain what type of data they are designed to model.

Feed-forward neural networks are designed to model structured data that is very predictable in structure, such as numerical data or categorical data that is one-hot encoded. CNNs are designed to model visual data, such as images or videos. RNNs are designed to model sequential data, such as textual data in the context of predicting what will come next; in other words, RNNs deal with data that is dependent on what data came before it.

## Problem 11.3 - Model Types

Your friend is starting to understand a little bit better. Now, in 2-3 sentences, explain how a CNN differs from an RNN. How do each of these differ from a "regular" or "feed-forward" neural network?

CNNs are neural networks that make use of patterns in neighboring pixels to make calculations and predictions on image data. RNNs are neural networks that make use of patterns in consecutive portions of data (often text) to make predictions about what data would follow (e.g. predicting the next word in a sentence). Both of these networks differ from regular neural networks because they analyze much more complex patterns in data, and therefore require not just a higher level of data preprocessing but also a different neural network structure in order to process the data.
